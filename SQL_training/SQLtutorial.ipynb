```python
from sqlalchemy import create_engine, inspect
#Syntax for create_engine is ("Dialect+Driver:////      "
engine = create_engine("sqlite+pysqlite://///workspaces/D2I-Jupyter-Notebook-Tools/SQL_training/chinook.db", echo=False) # Echo=True prints the SQL that the Python code is using

inspection = inspect(engine)
inspection.get_table_names()
```

```python
from sqlalchemy import MetaData, Table

# Create empty metadata class
metadata = MetaData()
```

```python
tracks = Table("tracks", metadata, autoload_with=engine)
#print(artists.name)
#print(artists.c.keys()) # print columns from table
print(repr(tracks)) # print info
```

```python
from sqlalchemy import text

with engine.connect() as con:
    statement = 'SELECT * FROM tracks' # Write a SQL statement
    results_proxy = con.execute(text(statement)) # Use text function to say we want to execute the SQL statement
    results = results_proxy.fetchall()
top_ten = results[:10]
print(top_ten)

for row in top_ten:
    print(row.Name)
```

```python
from sqlalchemy import select

with engine.connect() as con:
    statement = select(tracks)
    print(statement)
    results_proxy = con.execute(statement)
    results = results_proxy.fetchmany(size=10)
print(results)
```

```python
# Can convert straight to a dataframe

import pandas as pd

with engine.connect() as con:
    statement = select(tracks)
    results = con.execute(statement).fetchall()
df = pd.DataFrame(results)
df
```

```python
from sqlalchemy import or_ # SQL "or" in Python

with engine.connect() as con:
    statement = select(tracks).where(or_(tracks.columns.AlbumId == 1,
                                         tracks.columns.AlbumId == 2))
    results = con.execute(statement).fetchall()
df = pd.DataFrame(results)
df
```

```python
with engine.connect() as con:
    statement = select(tracks).order_by(tracks.columns.Composer.desc())
    results = con.execute(statement).fetchmany(10)
df = pd.DataFrame(results)
df
```

```python
# Order of length, get all tracks from AlbumId = 4 or 7
with engine.connect() as con:
    statement = select(tracks).where(or_(tracks.columns.AlbumId == 4, tracks.columns.AlbumId == 7)).order_by(tracks.columns.Milliseconds.asc())
    results = con.execute(statement).fetchall()
    print(statement)
df = pd.DataFrame(results)
df
```

```python
# Find summary statistics
from sqlalchemy import func

with engine.connect() as con:
    statement = select(func.sum(tracks.columns.Milliseconds))
    results = con.execute(statement).scalar()
print(results)
```

```python
from sqlalchemy import case, cast, Float

with engine.connect() as con:
    statement = func.sum(case((tracks.columns.Composer == 'roger glover', tracks.columns.UnitPrice),
                            else_=0))
    results = con.execute(statement).scalar()
    print(statement)
print(results)
```

```python
# total length of every song from albums (10 or 11) as a percentage of the total length of the library

with engine.connect() as con:
    statement_1 = select(func.sum(tracks.columns.Milliseconds))
    results_1 = con.execute(statement_1).scalar()

    statement_2 = func.sum(case((or_(tracks.columns.AlbumId == 10, tracks.columns.AlbumId == 11), tracks.columns.Milliseconds),
                                  else_=0))
    results_2 = con.execute(statement_2).scalar()

    print(statement_1)
    print(statement_2)
print(results_1)
print(results_2)
print((results_2 / results_1)*100)                    
                            
```

```python
albums = Table("albums", metadata, autoload_with=engine)

# Find columns in each table
print(albums.c.keys())
print(tracks.c.keys())
```

```python
# Join albums with tracks
with engine.connect() as con:
    statement = select(albums.columns.Title, tracks.columns.Name)
    statement.select_from(albums.join(tracks, albums.columns.AlbumId == tracks.columns.AlbumId))
    results = con.execute(statement).fetchmany(size=20)

print(statement)

df = pd.DataFrame(results)
df
```
